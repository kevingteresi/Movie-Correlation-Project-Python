```python
#IMPORT LIBRARIES

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configuration of the plots we will create
```

```python
# Read in the data

df = pd.read_csv(r'C:\Users\kevin\Downloads\movies.csv', encoding = "ISO-8859-1")
```

```python
# Look at the data

df.head()
```

```python
#Look for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data type for our columns

df.dtypes
```

```python
#convert some float rows to int to get rid of the .0's at the end

df['budget'] = df['budget'].astype('int')

df['gross'] = df['gross'].astype('int')
```

```python
df.head()
```

```python
# create correct yearcolumn

df['year_correct'] = df['released'].astype('str').str[0:4]
df['year_correct'] = df['year_correct'].astype('int')
df.head()
```

```python
#sort by gross revenus

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#set so above dataframe (and all dataframes) can be scrolled

pd.set_option('display.max_rows', None)
```

```python
# predictions
# budget will have a high correlation with gross
# Company will also have a high correlation with gross
```

```python
# scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')

plt.show
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

```python
#Lets start looking at correlation
```

```python
df.corr(method = 'pearson')  #Pearson (what we're using), kendall, spearman
```

```python
#lets get rid of inaccurate year column that was previously made
df = df.drop('year', axis=1)
df.head()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#looks at company
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
#new correlation matrix for new categorical variables

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
